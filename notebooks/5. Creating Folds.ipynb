```python
#importing the libraries
import pandas as pd
from sklearn import model_selection
```

```python
#reading the data
df = pd.read_csv("../data/train_engineered.csv")
```

```python
df.head()
```

```python
#creating a new column kfold
df.loc[:, 'kfold'] = -1
```

```python
#shuffling the data
df = df.sample(frac=1).reset_index(drop=True)
```

```python
df.head()
```

```python
y = df['Loan_Status'].values
```

```python
skf = model_selection.StratifiedKFold(n_splits=5)
```

```python
#creating the folds
for f, (t, v) in enumerate(skf.split(X=df, y=y)):
    df.loc[v, "kfold"] = f
```

```python
df.to_csv("../data/train_folds.csv", index=False)
```
